```python
import pandas as pd
import numpy as np
import requests
import csv
from bs4 import BeautifulSoup
```

```python
# Build the raw CIK_dict, CIK is the key of comany name.
import csv
CIK_dict={}
with open('cik-lookup-data.txt', 'r',encoding="ISO-8859-1") as csvfile:
    spamreader = csv.reader(csvfile, delimiter=':')
    for row in spamreader:
        CIK_dict[row[1]]=row[0]
    csvfile.close()
```

```python
print(len(CIK_dict))
```

```python
# Filter out those without 10-K from the dict
for CIK in list(CIK_dict):
    url="https://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK="+CIK+"&type=10-K"
    response = requests.get(url)
    if response.status_code == 200:
        print(CIK)
    else:
        print("Failure")
        break
    
    results_page = BeautifulSoup(response.content,'lxml')
    all_td_tags = results_page.find_all('td',nowrap="nowrap")
    if len(all_td_tags)==0:
        del CIK_dict[CIK]
```
